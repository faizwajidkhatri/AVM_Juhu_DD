Greating a 1D array

```python
x = [1,2,3,4,5,6]
y = np.array(x)
print(x)
print(y)
```

```python
print(type(x))
print(type(y))
```

Create a null array of size 10

```python

Z = np.zeros(10)
print(Z)
```

Create a null array of size 10 but the fifth value which is 1 

```python

Z = np.zeros(10)
Z[4] = 1
print(Z)
```

Reverse a vector (first element becomes last)

```python
Z = np.arange(50)
Z = Z[::-1]
print(Z)
```

Create a 3x3 matrix with values ranging from 0 to 8

```python
Z = np.arange(9).reshape(3, 3)
print(Z)
```

Create a 3x3x3 array with random values 

```python
Z = np.random.random((3,3,3))
print(Z)
```

Indexing Array

Creating a 2D array

```python
array = np.array([(1,2,3),(4,5,6)])
print(array)
```
